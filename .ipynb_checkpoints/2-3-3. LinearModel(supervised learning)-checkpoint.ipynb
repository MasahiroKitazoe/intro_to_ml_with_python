### 線形モデル

- 入力特徴量の線形関数（linear function）を用いて予測を行う

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mglearn
%matplotlib inline
from IPython.display import display
from sklearn.model_selection import train_test_split
```

```python
mglearn.plots.plot_linear_regression_wave()
```

### 線形回帰（通常最小二乗法）

- 最も単純で、最も古典的な線形回帰手法
- 訓練データにおいて、予測と真の回帰ターゲットyとの平均二乗誤差（mean squared error）が最小になるようにパラメータ　wとbを求める
- 平均二乗誤差 ＝ 予測都心の値との差を二乗したものの平均値
- 線形回帰にはパラメータがない。モデルの複雑さを制御する方法がない

```python

from sklearn.linear_model import LinearRegression
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
```

- 傾きを表すパラメータ（w）は、重み、もしくは係数（coefficient）と呼ばれ、coef_属性に格納される
- 切片(intercept, b)はintercept_属性に格納される

```python
print("lr.coef_: {}".format(lr.coef_))
print("lr.intercept_: {}".format(lr.intercept_))
```

```python
#訓練セットとテストセットに対する性能を見てみよう
print("Training set score: {:.2f}".format(lr.score(X_train, y_train)))
print("Training set score: {:.2f}".format(lr.score(X_test, y_test)))
```

```python
# boston_housingのデータを使って、線形回帰を試みる
from sklearn.datasets import load_boston
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
```

```python
# 精度を確認
print("Training set score: {:.2f}".format(lr.score(X_train, y_train)))
print("Training set score: {:.2f}".format(lr.score(X_test, y_test)))
```

- 訓練セットとテストセットで大きく性能が異なる　→　過剰適合が起きている
- 複雑度を制御できるモデルを探さなければならない
- リッジ回帰を使って見る

### リッジ回帰

- 予測に用いられる式は、通常最小二乗法と同じ
- ただし、係数（w）を訓練データに対する予測だけでなく、他の制約に対しても最適化する
- 係数（w）の大きさを可能な限り小さくしたい
    - 予測をうまく行いつつ、ここの特徴量が出力に与える影響を小さくしたい（つまり傾きを小さくしたい）
    - この制約条件は、正則化（regularization）の一例

### 正則化（regularization）

- 過剰適合を防ぐために明示的にモデルを制約すること
- リッジ回帰で用いられる正則化は、L2正則化と呼ばれる

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge.score(X_test, y_test)))
```

- LinearRegressionモデルよりもRidgeモデルの方が、過剰適合の危険が少なく、汎化性能が高い
- Ridgeモデルでは、モデルの簡潔さ（0に近い係数の数）と、訓練セットに対する性能がトレードオフになる
- 「このどちらに重きを置くのか」は、ユーザがalphaパラメータを用いて指定することができる
- alphaを増やすと、係数はより0に近くなり、訓練セットに対する性能は低下するが、汎化性能にとってはプラスになるかもしれない

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge10.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge10.score(X_test, y_test)))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge01.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge01.score(X_test, y_test)))
```

- alphaパラメータのモデルへの影響を定量的に知るには、様々なalphaに対するモデルのcoef_属性を調べる
- alphaが大きくなると、モデルがより制約されるので、alphaが大きい場合は小さい場合よりもcoef_要素の絶対値が小さくなることが期待される

```python
plt.plot(ridge.coef_, 's', label="Ridge alpha1")
plt.plot(ridge10.coef_, '^', label="Ridge alpha10")
plt.plot(ridge01.coef_, 'v', label="Ridge alpha01")

plt.plot(lr.coef_, 'o', label="LinearRegression")
plt.xlabel("Coefficient index")# 係数の番号
plt.ylabel("Coefficient magnitude")# 係数の大きさ

plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-25, 25)
plt.legend()
```

```python
# alphaの値を固定し、利用できる訓練データの量を変化させて見る
# boston_housingデータセットからサイズを大きくしながらデータを抽出して、LenearRegressionとRidge（alpha=1）で学習させた結果を評価したものを描画する
# この図は、モデルの性能をデータセットサイズの関数として示したもので、学習曲線と呼ばれる（learning curve）
mglearn.plots.plot_ridge_n_samples()
```

- データポイントが400以下の時は、線形回帰は何も学習することができていない
- しかし、データが増えるほど、二つとも性能が向上し、最後には線形回帰がリッジ回帰に追いつく
    →十分な訓練データがある時は、正則化はあまり重要でなくなる
- データが増えると、線形回帰のtrainingへの予測精度が下がっている。データ量が増えると、モデルが過剰適合しにくくなることがわかる

### Lasso

- Ridgeに代わる線形回帰の一種
- Ridge同様、係数が0になるように制約をかけるが、掛け方が少し違い、L1正則化と呼ばれる
- Lassoでは、いくつかの係数が完全に0になる
- モデルにおいて、いくつかの特徴量が完全に無視されることになる(自動的に特徴量を選択している)

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso.score(X_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso.coef_ !=0)))
```

- 適合不足を減らすために、alphaを減らすとよい
- この際、max_iter（最大繰り返し回数）をデフォルト値からふやしてやる必要がある

```python
# max_iterの値を増やしている
# こうしておかないと、モデルがmax_iterを増やすように警告する
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso001.score(X_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso001.coef_ !=0)))
```

```python
lasso0001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso0001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso0001.score(X_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso0001.coef_ !=0)))
```

```python
plt.plot(lasso.coef_, 's', label="Lasso alpha1")
plt.plot(lasso001.coef_, '^', label="Lasso alpha0.1")
plt.plot(lasso0001.coef_, 'v', label="Lasso alpha0.0001")

plt.plot(ridge01.coef_, 'o', label="Ridge alpha=0.1")
plt.legend(ncol=2, loc=(0, 1.05))
plt.ylim(-25, 25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
```

### クラス分類のための線形モデル

- 線形モデルによるクラス分類では、決定境界が入力の線形関数になる
- 線形の2クラス分類器は、2つのクラスを直線や平面や超平面で分割するということ
- 線形モデルを学習するためのアルゴリズム
    - 係数と切片の特定の組み合わせと訓練データの適合度を図る尺度
    - 正規化を行うか。行うならどの方法使うか
- 代表的な線形クラス分類アルゴリズム
    - ロジスティック回帰（logistic regression）
    - 線形サポートベクタマシン(linear support vector machines: SVM)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

X, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1, 2, figsize=(10, 3))

for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5,
                                   ax=ax, alpha=.7)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{}".format(clf._class_._name_))
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
axes[0].legend()
```

```python
mglearn.plots.plot_linear_svc_regularization()
```

- 左のプロットは小さいCを用いていて強力な正則化を行う場合に対応する
- Cが少し大きい真ん中のプロットは、モデルは2点の分類に失敗したサンプルに着目するので、決定境界が傾いている
- 一番右のプロットでは、決定きょうかの傾きはさらに急になり、クラス0の全ての点を正しくクラス分類できるようになる
- 一番右のモデルでは、全ての点を正しくクラス分類することに注力するあまり、クラス全体としてのレイアウトを捉えきれていない。(つまり、このモデルは過剰適合している)

```python
# 特徴量が多い場合の線形回帰
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test= train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42)
logreg = LogisticRegression().fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg.score(X_test, y_test)))
```

- 訓練セットとテストセットの精度がとても近い＝適合不足の可能性

```python
logreg100 = LogisticRegression(C=100).fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg100.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg100.score(X_test, y_test)))
```

- さらに正則化を強力にしていく

```python
logreg001 = LogisticRegression(C=0.01).fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg001.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg001.score(X_test, y_test)))
```

- 予想通り、適合不足だったモデルからさらに左側に寄ってしまい、訓練セットもテストセットも精度がデフォルトより悪化している

```python
plt.plot(logreg.coef_.T, 'o', label="C=1")
plt.plot(logreg100.coef_.T, '^', label="C=100")
plt.plot(logreg001.coef_.T, '^', label="C=0.01")
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
plt.hlines(0, 0, cancer.data.shape[1])
plt.ylim(-5, 5)
plt.xlabel("Feature")
plt.ylabel("Coefficient magnitude")
plt.legend()
```

- より解釈しやすいモデルが欲しいなら、わずかな特徴量しか使わないよう調節するL1正則化を使うよ良いかもしれない

```python
for C, marker in zip([0.001, 1, 100], ['o', '^', 'v']):
    lr_l1 = LogisticRegression(C=C, penalty="l1").fit(X_train, y_train)
    print("Training accuracy of l1 logreg with C={:.3f}: {:.2f}".format(
        C, lr_l1.score(X_train, y_train)))
    print("Test accuracy of l1 logreg with C={:.3f}: {:.2f}".format(
        C, lr_l1.score(X_test, y_test)))
    plt.plot(lr_l1.coef_.T, marker, label="C={:.3f}".format(C))
    
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
plt.hlines(0, 0, cancer.data.shape[1])
plt.xlabel("Feature")
plt.ylabel("Coefficient magnitude")

plt.ylim(-5, 5)
plt.legend(loc=3)
```

### 線形モデルによる多クラス分類

- 1対その他アプローチ（one-vs.-rest）

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state=42)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class0", "Class1", "Class2"])
```

```python
# LinearSVCクラス分類器をこのデータセットに学習させる
linear_svm = LinearSVC().fit(X, y)
print("Coefficient shape: ", linear_svm.coef_.shape)# 係数配列の形状
print("Intercept shape: ", linear_svm.intercept_.shape)# 切片配列の形状
```

```python
# 3つのクラス分類器による直線を可視化する
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, ['b', 'r', 'g']):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel("Feature 0")
plt.xlabel("Feature 1")
plt.legend(['Class 0', 'Class 1', 'Class 2', 'Line class 0', 'Line class 1', 'Line class 2'], loc=(1.01, 0.3))
```
